```python
import pandas as pd
import numpy as np
import json
```

```python
df= pd.read_html('https://www.liutaiomottola.com/formulae/freqtab.htm', header=0)
df= df[0]
```

```python
cols= df.columns.to_numpy()
cols_modified= cols[:3]
cols_modified= cols_modified[np.array([1,0,2])]
df= df.drop(columns= cols_modified[3:])
df= df[cols_modified]
df.head()
```

```python
df= df.dropna()
df= df.rename(columns={'Frequnecy (Hz)':'Frequency'})
df.head()
```

```python
df['Frequency (Hz)']= df['Frequency (Hz)'].round().astype('int')
df['Octave']= df['Octave'].astype('int')
df= df.reset_index(drop= True)
df.head(20)
```

```python
df_note= df[['Note Name']]
df_note= df_note.squeeze('columns')
df_note= df_note.replace(r'/...','', regex=True)
df_note= df_note.str.strip()
df['Note Name']= df_note
df= df.rename(columns={'Note Name':'Note', 'Frequency (Hz)':'Frequency'})
df.head(20)
```

```python
j = (df.groupby(['Octave'])
       .apply(lambda x: dict(zip(df.Note,x.Frequency)))
       .reset_index()
       .rename(columns={0:'Notes'})
       .to_json(orient='records'))


print(j)
```

```python
with open('frequency_chart.json', 'w') as data:
    json.dump(j, data, indent=4)
```
