```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_html('https://www.liutaiomottola.com/formulae/freqtab.htm', header=0)
df= df[0]
```

```python
cols= df.columns.to_numpy()
df= df.drop(columns= cols[3:])
df.head()
```

```python
df= df.dropna()
df.head()
```

```python
df['Frequency (Hz)']= df['Frequency (Hz)'].round().astype('int')
df['Octave']= df['Octave'].astype('int')
df= df.reset_index(drop= True)
df.head(20)
```

```python
df_note= df[['Note Name']]
df_note= df_note.squeeze('columns')
df_note= df_note.replace(r'/...','', regex=True)
df_note= df_note.str.strip()
df['Note Name']= df_note
df= df.rename(columns={'Note Name':'Note'})
df.head(20)
```

```python
df['Note']= df['Note'] + df['Octave'].astype('str')
df= df.drop(columns= 'Octave')
df.head(20)
```
